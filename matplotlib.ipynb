```python
#import for librarires


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt




```

```python
# axices

area = [100, 130, 145, 156, 167]

tower1= np.array([22332, 34333, 45335, 38337, 53389])
tower2= np.array([22235, 35643, 49855, 123387, 144589])
```

```python
#drawing the plot shape   (_Plot)


plt.title('Area vs Price ')

plt.plot(area, tower1, color = 'red', linewidth = 2, linestyle = '--')
plt.plot(area, tower2, color = 'blue', linewidth = 2)


plt.show()
```

```python

#drawing the subplot shape



plt.title('Area vs Price ')

            #row, Cols, num
plt.subplot(2, 1, 1)
plt.plot(area, tower1, color = 'red', linewidth = 2, linestyle = '--')


plt.subplot(2, 1, 2)
plt.plot(area, tower2, color = 'blue', linewidth = 2)


plt.show()

```

```python


#drawing the scatter shape   (Scatter)


```

```python
plt.scatter(area, tower1, color = 'red', linewidth = 2, marker='x', s = 40 )

plt.show()
```

```python
plt.scatter(area, tower2, color = 'blue', linewidth = 2, marker='x', s = 40 )

plt.show()
```

```python
#using generating random vaalues

x = np.random.rand(12)
y = np.random.rand(12)


plt.scatter(x, y, color = 'orange', marker='.', s=150)

plt.show()
```

```python

#drawing the histogram shape   (histogram)


data= np.array(['Engineer', 'Engineer', 'Doctor', 'Accountant', 'Accountant', 'Engineer', 'HR', 'Engineer', 'HR', 'HR'])


plt.hist(data, color ='green', histtype ='bar', alpha = .7, bins = 7)

plt.show()


```

```python

# 2D (histogram)


data= np.array([['Engineer', 'Engineer', 'Doctor', 'Accountant'], ['Engineer', 'Engineer', 'Accountant', 'Doctor']])


plt.hist(data, histtype ='bar', alpha = .7, bins = 7)

plt.show()

```

```python
#drawing the bar graph   (bar giraph)

a = np.array([1, 3, 5])
b = np.array([11, 13, 15])

plt.title('Bar Graph')
plt.bar(a, b, color = 'green')

plt.show()


```
